```python
import numpy as np
import pandas as pd
import bokeh as b

from bokeh.models import ColumnDataSource, CategoricalColorMapper
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.palettes import all_palettes

output_notebook()

# Importing the data
point_data = pd.read_csv("point_data.txt")

# Creating sources
source_point = ColumnDataSource(point_data)


#category = point_data['cluster'][0]
print(color_list.size)

palette = all_palettes['Dark2']

# Creating a factor color map
color_list = np.empty(shape = point_data['cluster'].size, dtype = "<U10")

for i in range(0,color_list.size):
    category = point_data.iloc[i,2]    
    color_list[i] = palette[6][category]    

# Add the color array to the ColumnDataSource
source_point.add(color_list, 'color')

# create a new plot with default tools, using figure
p = figure(plot_width=400, plot_height=400)

# add a circle renderer with a size, color, and alpha
p.circle(x = 'x', y = 'y', color = 'color', size=5, fill_alpha=0.5, source = source_point)

show(p) # show the results
```
